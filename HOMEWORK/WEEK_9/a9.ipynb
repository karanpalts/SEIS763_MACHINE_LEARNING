```python
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import classification_report, confusion_matrix


# 1. Load Data 

file_name = 'ML_HW_Data_CancerGene.xlsx'

X_train = pd.read_excel(file_name, sheet_name='geneexpTrain', header=None)
X_test = pd.read_excel(file_name, sheet_name='geneexpTest', header=None)

# keep_default_na=False prevents Pandas from turning the "NA" tumor class into a NaN float.
y_train = pd.read_excel(file_name, sheet_name='tumortypeTrain', header=None, keep_default_na=False).values.ravel()
y_test = pd.read_excel(file_name, sheet_name='tumortypeTest', header=None, keep_default_na=False).values.ravel()

# 2. Train the Decision Tree (Question 1)

clf = DecisionTreeClassifier(
    criterion='gini',
    max_depth=5,            
    min_samples_leaf=2,     
    random_state=42         
)
clf.fit(X_train, y_train)

# 3. Prediction & Evaluation (Question 2 & 3)

y_pred = clf.predict(X_test)

print("=== Confusion Matrix ===")
# The axes are [True Label, Predicted Label]
print(confusion_matrix(y_test, y_pred))

print("\n=== Classification Report ===")
# zero_division=0 handles the 0.00 metrics for the unseen NA class
print(classification_report(y_test, y_pred, zero_division=0))
```
