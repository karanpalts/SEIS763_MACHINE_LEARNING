```python
# 1. Load the patient data from “ML_HW_Data_Patients.csv” file. 

import pandas as pd

data = pd.read_csv("ML_HW_Data_Patients.csv")

# inspect the data to understand its structure and contents.

print(data.head()) # take note of gender, last name, and location

print(data.info())

print(data.describe())
```

```python
# # 2. Use variables Age, Gender, Height, Weight, Smoker, Location, SelfAssessedHealthStatus to build a linear regression model to predict the
# systolic blood pressure. You do NOT need to split data into training and testing sets.
import statsmodels.api as sm
target = "Systolic"
features = ["Age", "Gender", "Height", "Weight", "Smoker", "Location", "SelfAssessedHealthStatus"]

# create x (Features) and y (Target)
X = data[features].copy()
y = data[target]

# use drop_first=True to avoid dummy variable trap 
# this should remove one column per catergoy to act as that ref group

X = pd.get_dummies(X, columns=['Gender', 'Location', 'SelfAssessedHealthStatus'], drop_first=True)

# add intercept 
X = sm.add_constant(X)

# ensure all data is float to prevent those type errors
X = X.astype(float)

# train model
model = sm.OLS(y, X).fit()

# show results
print(model.summary())

```
